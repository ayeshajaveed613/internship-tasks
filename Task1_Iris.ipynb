```python
# -----------------------------------------
# TASK 1: Exploring and Visualizing Iris Dataset
# -----------------------------------------

# Step 1: Import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")
%matplotlib inline

# Step 2: Load dataset
df = sns.load_dataset('iris')

# Display first few rows
print("First 5 rows of the dataset:")
display(df.head())

# Step 3: Dataset overview
print("\nDataset Shape:", df.shape)
print("\nColumn Names:", df.columns.tolist())

print("\nDataset Info:")
df.info()

print("\nMissing Values:")
print(df.isnull().sum())

# Step 4: Summary statistics
print("\nSummary Statistics:")
display(df.describe().T)

# Step 5: Pairplot to visualize species separation
print("\nGenerating Pairplot...")
sns.pairplot(df, hue='species', corner=True)
plt.suptitle("Pairplot of Iris Dataset by Species", y=1.02)
plt.show()

# Step 6: Histograms for numeric features
print("\nGenerating Histograms...")
df.hist(figsize=(10,5), bins=12)
plt.suptitle("Histograms of Iris Features")
plt.show()

# Step 7: Boxplots for each numeric feature
print("\nGenerating Boxplots...")
num_cols = df.select_dtypes(include=np.number).columns

plt.figure(figsize=(12,4))
for i, col in enumerate(num_cols, 1):
    plt.subplot(1, 4, i)
    sns.boxplot(y=df[col])
    plt.title(col)

plt.tight_layout()
plt.show()

# Step 8: Correlation heatmap
print("\nGenerating Correlation Heatmap...")
corr = df.corr(numeric_only=True)

plt.figure(figsize=(6,4))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap of Iris Features")
plt.show()

# Step 9: Species-wise averages
print("\nSpecies-wise Average Measurements:")
display(df.groupby("species").mean())

```
